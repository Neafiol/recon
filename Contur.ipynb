```python
import tf_pose
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
with open("data.pk","rb") as f:
    photos = pickle.load(f)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
for photo in tqdm(os.listdir("data")[1472:]):
    try:
        coco_style = tf_pose.infer(f"data/{photo}")
        im = cv2.imread(f"data/{photo}")
        photos.append({
            "name":photo,
            "points":coco_style,
            "shape":im.shape
        })
    except:
        print("error",photo)
```
